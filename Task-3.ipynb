```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv("householdtask3.csv")
display(data.head(10))
```

```python
#Scatter plot
plt.scatter(data["year"],data["own"])
# Addingtitle to plot
plt.title("Scatter plot")
# Scattering x and y labels
plt.xlabel("year")
plt.ylabel("own")
# Showing the result
plt.show()
```

```python
#Line plot
plt.plot(data["year"])
plt.plot(data["own"])
# Addingtitle to plot
plt.title("Line plot")
# Scattering x and y labels
plt.xlabel("year")
plt.ylabel("own")
# Showing the result
plt.show()
```

```python
#Line plot
plt.bar(data["year"],data["own"])
# Addingtitle to plot
plt.title("Bar plot")
# Scattering x and y labels
plt.xlabel("year")
plt.ylabel("own")
# Showing the result
plt.show()
```

```python
#histogram
plt.hist(data["income"])
plt.title("Histogram")
plt.show()
```
